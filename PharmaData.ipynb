RFM ANALYSIS

```python
import datetime as dt
import pandas as pd #import pandas

recent_date = pd.to_datetime(str(df['Year'].max())) #Convert recent_date to datetime
rfm = df.groupby('Customer_ID').agg({
    'Year': lambda x: (recent_date - pd.to_datetime(x.astype(str).max())).days,  # Recency, Convert column 'x' to datetime within the lambda function
    'Sales_ID': 'count',  # Frequency
    'Sales': 'sum'  # Monetary
})
rfm.rename(columns={'Year': 'Recency', 'Sales_ID': 'Frequency', 'Sales': 'Monetary'}, inplace=True)
```

PRODUCT AFFINITY

```python
# Assuming 'ORDERID' and 'PRODUCT' columns represent transactions and purchased items
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder

# Create a list of transactions (each transaction is a list of products)
transactions = df.groupby('Sales_ID')['Product_ID'].apply(list).tolist()

# Prepare data for Apriori algorithm
te = TransactionEncoder()
te_ary = te.fit(transactions).transform(transactions)
transaction_df = pd.DataFrame(te_ary, columns=te.columns_)

# Apply Apriori and association rule mining
# Lower the min_support value to potentially find frequent itemsets
frequent_itemsets = apriori(transaction_df, min_support=0.01, use_colnames=True)

# Check if frequent_itemsets is empty and handle it
if frequent_itemsets.empty:
    print("No frequent itemsets found with the given min_support. Try lowering it further.")
else:
    # The `frequent_itemsets` DataFrame has a column named 'itemsets' which contains frozensets
    # We convert these frozensets to lists and get their lengths to determine the number of items in each itemset
    num_itemsets = frequent_itemsets['itemsets'].apply(lambda x: len(list(x))).max()

    # Pass 'num_itemsets' to the association_rules function to get association rules based on lift
    rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1, num_itemsets=num_itemsets)
    print(rules)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Specify the year you want to filter for
year_to_filter = 2020  # Replace with the desired year

# Filter the DataFrame for the specified year
filtered_df = df[df['Year'] == year_to_filter]

# Group by 'Product_ID' and calculate total sales for the filtered year
sales_by_product = filtered_df.groupby('Product_ID')['Sales'].sum().reset_index()

# Create the line plot
# In this case, since we are plotting for a single year, a bar plot might be more suitable
plt.figure(figsize=(10, 6))  # Adjust figure size as needed
sns.barplot(x='Product_ID', y='Sales', data=sales_by_product)
plt.title(f'Sales by Product for Year {year_to_filter}')
plt.xlabel('Product ID')
plt.ylabel('Total Sales')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to prevent labels from overlapping
plt.show()
```
